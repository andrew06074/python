```python
import pandas as pd
```

```python
c_2015 = pd.read_csv('combine_2015.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c_2015
```

```python
c_2015 = c_2015.replace(np.nan,0)
```

```python
player_fixed = c_2015.Player
```

```python
player_fixed = player_fixed.str.split('\\', n = 1, expand = True)
```

```python
c_2015['Name'] = player_fixed[0]
```

```python
c_2015.drop(columns=['Player'], inplace = True)
```

```python
c_2015.columns = ['Pos', 'School', 'College', 'Ht', 'Wt', 'Fourty_yd', 'Vertical', 'Bench',
       'Broad_jump', 'Three_cone', 'Shuttle', 'Drafted', 'Name']
```

```python
c_2015['Total'] = c_2015.Fourty_yd + c_2015.Vertical + c_2015.Bench + c_2015.Broad_jump + c_2015.Three_cone
```

```python
draft = c_2015.Drafted
```

```python
draft = draft.str.split(' / ', n=4,expand = True)
```

```python
draft
```

```python
c_2015['Round'] = draft[1]
```

```python
c_2015['Pick'] = draft[2]
```

```python
c_2015_new = c_2015[['Name','Pos','Fourty_yd','Vertical','Bench','Broad_jump','Shuttle','Total','Round','Pick']]
```

```python
c_2015_new
```

```python
#strip text out of round column
new_rnd = c_2015_new.Round
new_rnd = new_rnd.str.split('s', n=1, expand = True)
new_rnd = new_rnd[0]
new_rnd = new_rnd.str.split('n', n=1, expand = True)
new_rnd = new_rnd[0]
new_rnd = new_rnd.str.split('r', n=1, expand = True)
new_rnd = new_rnd[0]
new_rnd = new_rnd.str.split('t', n=1, expand = True)
new_rnd = new_rnd[0]
c_2015_new.Round = new_rnd
```

```python
#strip text out of pick column
new_pick = c_2015_new.Pick
new_pick = new_pick.str.split('s', n=1, expand = True)
new_pick = new_pick[0]
new_pick = new_pick.str.split('n', n=1, expand = True)
new_pick = new_pick[0]
new_pick = new_pick.str.split('r', n=1, expand = True)
new_pick = new_pick[0]
new_pick = new_pick.str.split('t', n=1, expand = True)
new_pick = new_pick[0]
c_2015_new.Pick = new_pick
```

```python
c_2015_new
```

```python
c_2015_new.Round = c_2015_new.Round.astype(float)
c_2015_new.Pick = c_2015_new.Pick.astype(float)
c_2015_new = c_2015_new.sort_values(by=['Round', 'Pick'])
```

```python
round_1 = c_2015_new.sort_values(by=['Round', 'Pick'])
```

```python
overall_pick = round_1.Pick
overall_pick.reset_index(drop=True,inplace=True)
overall_pick.reset_index()
c_2015_new['Overall_pick'] = overall_pick.index + 1
c_2015_new
```

```python
c_2015 = c_2015_new
```

```python
c_2015.reset_index(drop=True,inplace=True)
```

```python
c_2015.to_csv('2015_draft_order', index=False)
```
